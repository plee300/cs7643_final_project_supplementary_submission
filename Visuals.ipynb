```python
import numpy as np

data = np.load('results/VisualDQN_eval_data.npy')
title = "DQN (3 Fully Connected Layers)"
```

```python
x = data[:, 0]
y = data[:, 1]
```

```python
import matplotlib.pyplot as plt
#running average for better visual
avg = 2
y = np.convolve(y, np.ones(avg)/avg, mode='valid')

plt.plot(x[:len(y)], y)
plt.xlabel("Episode")
plt.ylabel(f"Running Average Per-Episode\nReward over {avg*8} Episodes")
plt.title(title)
plt.show()
```
