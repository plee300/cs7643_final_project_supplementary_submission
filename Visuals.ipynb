```python
import numpy as np

#decay=2000
data1 = np.load('results/CTQN_train_data_10000.npy')
data2 = np.load('results/CTQN_train_data.npy')
data_2000 = np.concat((data1, data2))

#decay=200
data3 = np.load('results/CTQN_train_data_fixed_2000.npy')
data4 = np.load('results/CTQN_train_data_fixed_4000.npy')
data5 = np.load('results/CTQN_train_data_fixed_6000.npy')
data_200 = np.concat((data3, data4, data5))
```

```python
import matplotlib.pyplot as plt

x_200 = [i*data_200.shape[1] for i in range(data_200.shape[0])]
y_200 = np.average(data_200, axis=1)

#running average for better visual
avg = 10
y_200 = np.convolve(y_200, np.ones(avg)/avg, mode='valid')
```

```python
x_2000 = [i*data_2000.shape[1] for i in range(data_2000.shape[0])]
y_2000 = np.average(data_2000, axis=1)

y_2000 = np.convolve(y_2000, np.ones(avg)/avg, mode='valid')
```

```python
plt.plot(x_200[:len(y_200)], y_200, label="d=200")
plt.plot(x_2000[:len(y_2000)], y_2000, label="d=2000")
plt.xlabel("Episode")
plt.ylabel(f"Running Average Per-Episode\nReward over {avg*8} Episodes")
plt.title("DTQN (Transformer Sequence Memory)")
plt.legend()
plt.show()
```
